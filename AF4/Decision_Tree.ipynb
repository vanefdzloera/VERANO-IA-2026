# 1. Load Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

# 2. Load Data

```python
cancer_data = load_breast_cancer()
X = X = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
y = cancer_data.target

len(X)
```

# 3. Data cleaning

```python
print(X.isnull().sum())
```

# 4. Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

# 5. Train model

```python
modelo_arbol = DecisionTreeClassifier(max_depth=3, random_state=42)
modelo_arbol.fit(X_train, y_train)
```

```python
train_accuracy = modelo_arbol.score(X_train, y_train)
print(train_accuracy)
```

# 6. Model Testing and Visualization

```python
y_pred = modelo_arbol.predict(X_test)
```

```python
#Prediction 1
print(y_pred[-1])
print(y_test[-1])
```

```python
#Prediction 2
print(y_pred[20])
print(y_test[20])
```

```python
#Prediction 3
print(y_pred[-1])
print(y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=cancer_data.target_names))
```

```python
plt.figure(figsize=(20, 10))
plot_tree(modelo_arbol,
          filled=True,
          feature_names=cancer_data.feature_names,
          class_names=cancer_data.target_names)
plt.title("Decision Tree for Breast Cancer Diagnosis")
plt.savefig("decision_tree.png", dpi=300)
plt.show()
```
